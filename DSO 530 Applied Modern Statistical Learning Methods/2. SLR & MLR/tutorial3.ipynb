```python
import pandas as pd
housing = pd.read_csv('housing.csv')
display(housing.head())
```

```python
import statsmodels.formula.api as smf
result1 = smf.ols('medv ~ crim', data = housing).fit()
result1.summary()
```

```python
result2 = smf.ols('medv ~ crim+rm',data = housing).fit()
result2.summary()
```

```python
string_cols = ' + '.join(housing.columns[:-1])
result3 = smf.ols(f'medv ~ {string_cols}',data = housing).fit()
result3.summary()
```

```python
string_cols = ' + '.join(housing.columns[:-1].difference(['zn']))
result4 = smf.ols(f'medv ~ {string_cols}',data = housing).fit()
result4.summary()
```

###### If interaction terms are used then the individual terms are included in the model by default

```python
result5 = smf.ols('medv ~ river*rm',data = housing).fit()
result5.summary()
```

```python
import numpy as np
result6 = smf.ols('medv ~ rm + np.power(rm,2)',data = housing).fit()
result6.summary()
```

```python
result1 = smf.ols('medv ~ rm',data = housing).fit()
result1.summary()
```

### Predicting data

```python
test_data = {'rm':[5,10,15,20]}
test_data_df = pd.DataFrame(test_data)
test_data_df
```

```python
prediction1 = result1.get_prediction(test_data_df)
prediction1.summary_frame(alpha = 0.05)
# the output is not CI, infact it is PI but abuse of language
```
